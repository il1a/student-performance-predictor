```python
# Import Packages
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import pandas as pd
```

```python
# URL to the raw content of the CSV file
url = 'https://raw.githubusercontent.com/il1a/AIBAS-student-performance-predictor/main/data/StudentPerformanceFactors.csv'

# Make an HTTP GET request to fetch the raw CSV content
response = requests.get(url)

if response.status_code == 200:
    # Convert the raw text to a pandas DataFrame
    from io import StringIO

    csv_data = StringIO(response.text)
    df = pd.read_csv(csv_data)

    print("Data scraped successfully! Here's the DataFrame information:")
    df.info()
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python
# Check the number of NaN values
total_na = df.isna().sum().sum()
print("\nTotal number of NaN values in the dataset:", total_na)
```

```python
# Drop NaN values and examine afterward
df = df.dropna()

total_na = df.isna().sum().sum()
print("\nTotal number of NaN values in the dataset:", total_na)
```

```python
# Visual check for outliers using boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, orient='v')
plt.title("Boxplot Before Removing Outliers")
plt.show()
```
